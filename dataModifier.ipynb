```python
import pandas as pd
```

```python
df= pd.read_csv('Hikes_averages.csv')
```

```python
#creat me a country named all which has the average of all countries except the name of course
average_row = df.mean(numeric_only=True)

# Create a new DataFrame row with the averages
average_row['country_name'] = 'All'  # Label the row
average_df = pd.DataFrame([average_row])

# Append the new row with averages to the original DataFrame
df_with_average = pd.concat([df, average_df], ignore_index=True)
```

```python
df = df_filtered
```

```python
grouped_averages = df.groupby('country_name').mean(numeric_only=True).reset_index()

# Calculate overall averages
overall_averages = df.mean(numeric_only=True).to_frame().T  # Transpose to make it a row
overall_averages['country_name'] = 'All'
```

```python
final_table = pd.concat([grouped_averages, overall_averages], ignore_index=True)
```

```python
final_table = final_table.drop(columns=['Unnamed: 0'])
```
